Q1] Define Power BI and What are the key components of the Power BI ecosystem? Briefly
explain:
● Power BI Desktop
● Power BI Service
● Power BI Mobile
● Power BI Gateway

Ans]

Power BI is a data visualization and business intelligence tool developed by Microsoft that helps users convert raw data into meaningful insights using interactive reports and dashboards.

Key Components of Power BI Ecosystem:

1 Power BI Desktop

A free application used to create reports and dashboards on your computer. It allows data cleaning, modeling, and visualization.

2 Power BI Service

An online (cloud-based) platform where reports are published, shared, and collaborated with others.

3 Power BI Mobile

A mobile app that lets users view and interact with reports and dashboards on smartphones and tablets.

4 Power BI Gateway

A tool that connects on-premises data sources (like local servers) to the Power BI Service for automatic data updates.

Q2] Compare the following Power BI visuals:

● Pie Chart vs Donut Chart

● Bar Chart vs Column Chart

When would you prefer one over the other? Give one example for each pair.

Ans]

1] Pie Chart vs Donut Chart:


Pie Chart:

Shows parts of a whole as slices. Simple and easy to understand when categories are few.


Donut Chart:

Similar to pie chart but with a hole in the center. Can display additional information (like total value) in the middle.

- When to use:

Use Pie Chart when you want a simple comparison of proportions.


Use Donut Chart when you also want to show extra info (like total).

- Example:

Showing market share of 4 companies → Pie Chart
Showing same data with total sales in center → Donut Chart

2] Bar Chart vs Column Chart:

Bar Chart:

 Displays data horizontally. Good for comparing many categories or long names.


Column Chart:

 Displays data vertically. Good for showing changes over time.

- When to use:

Use Bar Chart when category names are long or there are many items.
Use Column Chart when showing trends over time.

- Example:

Comparing sales of 10 products with long names → Bar Chart
Showing monthly sales growth → Column Chart

Q3] Explain the significance of:

● Star schema vs Snowflake schema

● Primary key vs Foreign key in relationships (Power BI)

Why is cardinality important?

Ans]

1] Star Schema vs Snowflake Schema

Star Schema:

A simple structure where one central fact table is directly connected to multiple dimension tables.

-  Significance: Easy to understand, faster performance, best for Power BI reports.
Snowflake Schema:

A complex structure where dimension tables are further divided into sub-tables.
-  Significance: Reduces data redundancy but is harder to manage and slightly slower.

2] Primary Key vs Foreign Key (in Power BI relationships)

Primary Key:

A unique column in a table (no duplicates). It identifies each record.

Foreign Key:

A column in another table that connects to the primary key. It is used to create relationships.

- Significance:
These keys help link tables correctly so Power BI can combine and analyze data properly.

3] Why is Cardinality Important?

Cardinality defines how tables are related (like one-to-one, one-to-many, many-to-many).

- Importance:

Ensures correct data relationships
Prevents duplicate or wrong results
Improves report accuracy and performance

- Example:
One customer can have many orders → One-to-Many relationship

Q4] Differentiate between:

- Calculated column vs Measure
Also, define Row context and Filter context with simple examples.

Ans]

1] Calculated Column vs Measure

Calculated Column:

Created for each row in a table using a formula. Stored in the data model.
- Used when you need row-by-row calculations.

Measure:

Calculated on the fly based on filters in reports. Not stored.
- Used for aggregations like SUM, AVERAGE.

- Example:

Calculated Column → Profit = Sales − Cost (for each row)
Measure → Total Sales = SUM(Sales)

2] Row Context

Row context means calculation is done row by row.

- Example:
If you create a calculated column:
Profit = Sales − Cost
It calculates profit for each individual row.

3] Filter Context

Filter context means calculation changes based on filters applied (like slicers, visuals).

- Example:
If you create a measure:
Total Sales = SUM(Sales)

Without filter → shows total sales
With filter (e.g., Year = 2024) → shows only 2024 sales

Q5] What is the difference between a report and a dashboard in Power BI?

Ans]

Report vs Dashboard in Power BI

Report:

A multi-page document with detailed visuals (charts, tables, graphs).
-  Used for deep analysis and exploration with filters and slicers.

Dashboard:

A single-page view that shows key information (KPIs) using tiles.
- Used for quick overview and decision-making.
